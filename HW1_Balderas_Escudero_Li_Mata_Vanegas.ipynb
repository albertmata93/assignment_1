```python
#function that will be used to calculate the computers choice of rock paper scissors
def prs(compbrains): 
        # use computer brains if any play was chosen by the human half of the time or more
        if compbrains['Rock'] >= .5*(compbrains['Rock']+compbrains['Paper']+compbrains["Scissors"]):
            comp_choice = 'Paper'
        elif compbrains['Paper'] >= .5*(compbrains['Rock']+compbrains['Paper']+compbrains["Scissors"]):
            comp_choice = 'Scissors'
        elif compbrains['Scissors'] >= .5*(compbrains['Rock']+compbrains['Paper']+compbrains["Scissors"]):
            comp_choice = 'Rock'
        #otherwise randomly select computer play
        else:
            comp_choice = random.choice(['Rock','Paper','Scissors'])
        
        
        return (comp_choice)
```

```python
#function that will be used to decide the winner of the game.
#returns true if human wins, false otherwise
def who_won(user_choice,comp_choice):
    if user_choice==comp_choice:
        print "You and the computer tied!"
        return False
    elif user_choice =="Rock" and comp_choice=='Scissors':
        print "Rock Smashes Scissors, YOU WIN!"
        return True
    elif user_choice =="Rock" and comp_choice=='Paper':
        print "Paper covers Rock, YOU LOSE!"
        return False
    elif user_choice =="Paper" and comp_choice=='Scissors':
        print "Scissors cuts Paper, YOU LOSE"
        return False
    elif user_choice =="Paper" and comp_choice=='Rock':
        print "Paper covers Rock, YOU WIN!"
        return True
    elif user_choice =="Scissors" and comp_choice=='Rock':
        print "Rock  smashes scissors, YOU LOSE!"
        return False
    elif user_choice =="Scissors" and comp_choice=='Paper':
        print "Scissors cuts Paper, YOU WIN!"
        return True
    return False
    
```

```python
#Probelm #3
#Song-writing.p
#This program fills in the XXX in a song with rhymed endings and complementary beats.
#Group 6
#Last Edit: Feb 14th

def main():

	#read the endings file
	endings = open("endings.txt","r")

	#create a dictionary
	end_dict = {}
	#create a reverse dictionary
	rev_dict = {}

	#print endings
	#readline and build library
	for word in endings:
		word = word.strip()
		word = word.split("::")
		end_dict[word[0]] = word [1:]
		rev_key = word[0], word[-1]
		
		if word[1] not in rev_dict:
			rev_dict[word[1]] = (rev_key)
		else:
			rev_dict[word[1]] += (rev_key)
	
	
	#read the skeleton
	poem = open("skeleton_SamuraiShowdown.txt","r")
	new_poem = ""
	endings.close()
	
	#read each line of the skeleton
	for line in poem:
		line = line.strip()
		line = line.split("::")
		count = int(line[1])
		
		#handling the no ending line
		if line[-1] != "XXX":
			if line[-1]== "":
				new_poem += line[0]
				new_poem += "\n"
			#set the rhyme and full length for the next line
			else:
				ending = line[-1]
				info = end_dict[ending]
				full_len = count + int(info[-1])
				rhyme = info[0]
			
			#add non XXX line to the new string
				# eliminating empty spaces
				if line[0] != "":
					new_poem += line[0]
					new_poem += " "
				new_poem += line[-1]
				new_poem += "\n"

		#replacing XXX line
		else:
			need = full_len - count
			
			options = rev_dict[rhyme]
			for item in options:
				if str(need) == item:
					index = options.index(item)
					line[-1] = options[index-1]
					new_poem += line[0]
					new_poem += " "
					new_poem += line[-1]
					new_poem += "\n"
	#close skeleton	
	poem.close()	
	#print the output
	print new_poem
	
main()
```

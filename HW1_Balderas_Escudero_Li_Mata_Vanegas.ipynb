```python
#Problem 1: Rock Paper Scissors
#We used a while loop to continuously run the program until the user inputted "Quit"
#We created the compbrains dictionary to store user inputs so the computer could respond to user biases
#The prs function was used to select the computer's choice either based on the user's biases or randomly
#if the user selected a certain option 50% or more of the time, the computer would play the optimal choice
#The who_one function determines the winner by comparing the user choice and the computer choice
#Everytime the games runs, the computer selects its choice and the user inputs their choice
#the who_won function is then used to determine a winner, and the totalgames & totalwins variables are incremented accordingly
# the computer's & user's choices are printed along with the result of the match
import random #we need this to randomly select the rock paper or scissors
print "Hello, Human \nWelcome to Rock,Paper, Scissors \nPlease enter your choice of 'Rock', 'Paper','Scissors', or 'Quit'"

#create flag to signal stop of program
quit = False

#variable to record number of games
totalgames = 0

#variable to record human wins
totalwins = 0

#dictionary to track user's choice
compbrains = {'Rock':0,'Paper':0,'Scissors':0}

#run program until user says "Quit"
while not quit:
    return_val = prs(compbrains) #this runs the function and saves the return values in a tuple called return_val
    comp_choice = return_val # this part assigns the first element of the tuple to computer choice
    user_choice = raw_input() 
    print "Comps: " + str(comp_choice)
    print "user: " + str(user_choice)

    if user_choice=='Rock' or user_choice=='Paper' or user_choice=='Scissors':
        compbrains[user_choice]+=1
        win = who_won(user_choice, comp_choice)
        if win:
            totalwins += 1
        totalgames +=1
    elif user_choice=="Quit":
        print "\nYou won " + str(totalwins) + " out of " + str(totalgames) + " games."
        print "Thanks for playing, see you soon!"
        quit = True
    else:
        print "Invalid input, please try again"  
```

```python
#function that will be used to calculate the computers choice of rock paper scissors
#inputs: dictionary of user's past choices
#outputs: the computer's choice
def prs(compbrains): 
        # use computer brains if any play was chosen by the human half of the time or more
        if compbrains['Rock'] >= .5*(compbrains['Rock']+compbrains['Paper']+compbrains["Scissors"]):
            comp_choice = 'Paper'
        elif compbrains['Paper'] >= .5*(compbrains['Rock']+compbrains['Paper']+compbrains["Scissors"]):
            comp_choice = 'Scissors'
        elif compbrains['Scissors'] >= .5*(compbrains['Rock']+compbrains['Paper']+compbrains["Scissors"]):
            comp_choice = 'Rock'
        #otherwise randomly select computer play
        else:
            comp_choice = random.choice(['Rock','Paper','Scissors'])
        
        
        return (comp_choice)
```

```python
#function that will be used to decide the winner of the game.
#inputs: the user's choice and the computer's choice
#outputs: prints the result of the game, returns true if human wins, false otherwise
def who_won(user_choice,comp_choice):
    if user_choice==comp_choice:
        print "You and the computer tied!"
        return False
    elif user_choice =="Rock" and comp_choice=='Scissors':
        print "Rock Smashes Scissors, YOU WIN!"
        return True
    elif user_choice =="Rock" and comp_choice=='Paper':
        print "Paper covers Rock, YOU LOSE!"
        return False
    elif user_choice =="Paper" and comp_choice=='Scissors':
        print "Scissors cuts Paper, YOU LOSE"
        return False
    elif user_choice =="Paper" and comp_choice=='Rock':
        print "Paper covers Rock, YOU WIN!"
        return True
    elif user_choice =="Scissors" and comp_choice=='Rock':
        print "Rock  smashes scissors, YOU LOSE!"
        return False
    elif user_choice =="Scissors" and comp_choice=='Paper':
        print "Scissors cuts Paper, YOU WIN!"
        return True
    return False
    
```

```python
#Problem #3
#Song-writing.py
#This program fills in the XXX in a song with rhymed endings and complementary beats.
#Group 6
#Last Edit: Feb 14th
#The program reads in the endings.txt file
#for each line in the file, the line is stripped and split to create a list
#each line in the file, a dictionary entry is created with the ending word as the key and the rhyming sound & beat as values
#rev_dict is created with the rhyming sound as the key and the tuple, rev_key, created with the ending word and the beat as the value. 
#if the rhyming sound for a new word already exist in the rev_dict dictionary, the new ending word and the beat will be appended to the exisitng value.
#The program opens the skeleton_SamuraiShowdown.txt file and saves it as poem
#for every line in poem, the beat number is stored in count
#if the line doesn't have a ending word, it is added to the new poem with a newline character
#if the line has a ending word, we store the prefix beats in the count and get the number of beats in the ending word from the end_dict dictionary and add together for the full_length.
#We store the rythme of ending word from the non-XXX line in variable rhyme
#If the line ends with XXX, we use the stored rhtyme as key to look for word options from the rev_dict, locate the index of the number of beats that adds up to full length with the prefix beat
#find the suitable word by using index -1 and replace XXX with the suitable word. The replaced line will be concatenated to the new_poem string with a new line character at the end, ready for output.
#funtion inputs: none
#function outputs: new_poem
def main():

	#read the endings file
	endings = open("endings.txt","r")

	#create a dictionary
	end_dict = {}
	#create a reverse dictionary
	rev_dict = {}


	#readline and build library
	for word in endings:
		word = word.strip()
		word = word.split("::")
		end_dict[word[0]] = word [1:]
		rev_key = word[0], word[-1]
		
		if word[1] not in rev_dict:
			rev_dict[word[1]] = (rev_key)
		else:
			rev_dict[word[1]] += (rev_key)
	
	
	#read the skeleton
	poem = open("skeleton_SamuraiShowdown.txt","r")
	new_poem = ""
	endings.close()
	
	#read each line of the skeleton
	for line in poem:
		line = line.strip()
		line = line.split("::")
		count = int(line[1])
		
		#handling the no ending line
		if line[-1] != "XXX":
			if line[-1]== "":
				new_poem += line[0]
				new_poem += "\n"
			#set the rhyme and full length for the next line
			else:
				ending = line[-1]
				info = end_dict[ending]
				full_len = count + int(info[-1])
				rhyme = info[0]
			
			#add non XXX line to the new string
				# eliminating empty spaces
				if line[0] != "":
					new_poem += line[0]
					new_poem += " "
				new_poem += line[-1]
				new_poem += "\n"

		#replacing XXX line
		else:
			need = full_len - count
			
			options = rev_dict[rhyme]
			for item in options:
				if str(need) == item:
					index = options.index(item)
					line[-1] = options[index-1]
					new_poem += line[0]
					new_poem += " "
					new_poem += line[-1]
					new_poem += "\n"
	#close skeleton	
	poem.close()	
	#print the output
	print new_poem
	
main()
```
